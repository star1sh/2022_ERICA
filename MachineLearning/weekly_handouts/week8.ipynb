```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head = 5
```

```python
plt.figure(figsize = (8,6))
plt.ylim(0,10) #y값의 최대 값 지정.(제시된 표와 같은 모양으로 만들기위해)
sns.boxplot(x = "day", y = "tip", hue ="time", data = tips) #x축에 요일, y축에 팁 표시, 구분 = 시간
plt.show()
```

```python
sns.set_style("darkgrid") #제시된 표와 같은 모양으로 만들기 위해 삽입한 코드
plt.figure(figsize = (8,6))
plt.ylim(0,50) #y값의 최대 값 지정.(제시된 표와 같은 모양으로 만들기 위해)
sns.boxplot(x = "size", y = "total_bill", hue ="day", data = tips) #x축에 손님수, y축에 전체지불금액표시, 구분 = 요일
plt.show()
```
